```python
df.loc[df['Type_red']==1]['ratings'].mean()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['Red', 'White', 'Sparkling', 'Rose', 'Dessert', 'Fortified']
mean_rating = [
    df.loc[df['Type_red']==1]['ratings'].mean(),
    df.loc[df['Type_white']==1]['ratings'].mean(),
    df.loc[df['Type_sparkling']==1]['ratings'].mean(),
    df.loc[df['Type_rose']==1]['ratings'].mean(),
    df.loc[df['Type_dessert']==1]['ratings'].mean(),
    df.loc[df['Type_fortified']==1]['ratings'].mean()
]
ax.set(ylim=[3.5, 5])
ax.bar(langs,mean_rating)
plt.show()



```

```python
data = [
    [
        df.loc[df['Type_red']==1]['Bold'].mean(),
        df.loc[df['Type_red']==1]['Tannic'].mean(),
        df.loc[df['Type_red']==1]['Sweet'].mean(),
        df.loc[df['Type_red']==1]['Alcohol content'].mean()
    ],
       [
        df.loc[df['Type_white']==1]['Bold'].mean(),
        df.loc[df['Type_white']==1]['Tannic'].mean(),
        df.loc[df['Type_white']==1]['Sweet'].mean(),
        df.loc[df['Type_white']==1]['Alcohol content'].mean()
    ],
       [
        df.loc[df['Type_sparkling']==1]['Bold'].mean(),
        df.loc[df['Type_sparkling']==1]['Tannic'].mean(),
        df.loc[df['Type_sparkling']==1]['Sweet'].mean(),
        df.loc[df['Type_sparkling']==1]['Alcohol content'].mean(),
    ],
       [
        df.loc[df['Type_rose']==1]['Bold'].mean(),
        df.loc[df['Type_rose']==1]['Tannic'].mean(),
        df.loc[df['Type_rose']==1]['Sweet'].mean(),
        df.loc[df['Type_rose']==1]['Alcohol content'].mean(),
    ],
       [
        df.loc[df['Type_dessert']==1]['Bold'].mean(),
        df.loc[df['Type_dessert']==1]['Tannic'].mean(),
        df.loc[df['Type_dessert']==1]['Sweet'].mean(),
        df.loc[df['Type_dessert']==1]['Alcohol content'].mean(),
    ],
       [
        df.loc[df['Type_fortified']==1]['Bold'].mean(),
        df.loc[df['Type_fortified']==1]['Tannic'].mean(),
        df.loc[df['Type_fortified']==1]['Sweet'].mean(),
        df.loc[df['Type_fortified']==1]['Alcohol content'].mean(),
    ]
]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'r', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'b', width = 0.25)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['contain eggs&milk', 'contain eggs', 'contain milk', 'neither']
mean_rating = [
    df.loc[(df['contain_eggs']==1)&(df['contain_milk']==1)]['ratings'].mean(),
    df.loc[(df['contain_eggs']==1)&(df['contain_milk']==0)]['ratings'].mean(),
    df.loc[(df['contain_eggs']==0)&(df['contain_milk']==1)]['ratings'].mean(),
    df.loc[(df['contain_eggs']==0)&(df['contain_milk']==0)]['ratings'].mean()
]
ax.set(ylim=[3.5, 4.5])
ax.bar(langs,mean_rating)
plt.show()
```
