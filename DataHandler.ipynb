```python
df['contain_eggs'] = df.apply(lambda row: 1 if ' egg allergens' in row.Allergens.split(',') else 0, axis=1)
df['contain_milk'] = df.apply(lambda row: 1 if ' milk allergens' in row.Allergens.split(',') else 0, axis=1)
df.drop(columns = ['Allergens'], inplace=True)
```
